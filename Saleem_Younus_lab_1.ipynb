## Exercise 3

1. Find and plot on the $x-$interval $[\pi,3.3]$: $\ln(1+9(x-\pi))$, $P_1(x)$, $P_2(x)$, $P_3(x), P_4(x)$, where the Taylor polynomials are centered at $\pi$.
2. Which one is more accurate when you get close to $\pi$? What happens when you get away from $\pi$?

````python
import numpy as np
from numpy import log as ln

pi = np.pi 
x = np.linspace(pi , 3.3, 100)
y = ln(9*(x - pi) + 1)
plt.plot(x, y, label = "f (x)", color = "black")

a = 9*(x - pi)      
b = 9*(x - pi) - (81/2)*(x - pi)**2 
c = 9*(x - pi) - (81/2)*(x - pi)**2 + 243*(x - pi)**3     
d = 9*(x - pi) - (81/2)*(x - pi)**2 + 243*(x - pi)**3 - (6561/4)*(x - pi)**4

plt.plot(x, a, label = "P1(x)")
plt.plot(x, b, label = "P2(x)") 
plt.plot(x, c, label = "P3(x)")
plt.plot(x, d, label = "P4(x)")

plt.legend()
plt.show()

plt.plot(x, np.abs(y-a), label = "P1(x)")
plt.plot(x, np.abs(y-b), label = "P2(x)")
plt.plot(x, np.abs(y-c), label = "P3(x)")
plt.plot(x, np.abs(y-d), label = "P4(x)")

plt.legend()
plt.show()

print('P4(x) most accuratly estimates f(x) when you get close to . When you get away from , P2(x) has highest accuracy ')
````

## Exercise 5
Let $f(x)=\sqrt[4]{x}$
1. Find and plot $P_3(x)$ centered at $a=1$
2. Plot $P_3(x)$ and $f(x)$ for in $x\in[0,4]$
3. Plot the error between $P_3$ and $f$
3. Approxmation $\sqrt[4]{2}$ and $\sqrt[4]{3}$  using $P_3$. How far off are they?

````python
x = np.linspace(0, 4)

def f(x):
  return x**(1/4)

def P3(x):
  return 1 + (x - 1) / 4 + ((-3)*(x - 1)**2) / 32 + ((21)*(x - 1)**3) / 128


plt.plot(x, f(x), label = "f(x)")
plt.plot(x, P3(x), label = "P3(x)")
plt.legend()

plt.show()
plt.plot(x, np.abs(f(x) - P3(x)), label='Error: |P3(x) - f(x)|')
plt.legend()
plt.show()

print("Using P3, 2^(1/4) is approximated as " + str(P3(2)) + ". This is " + str(np.abs(P3(2) - f(2)))+ " off from the exact value")
print("Using P3, 3^(1/4) is approximated as " + str(P3(3)) + ". This is " + str(np.abs(P3(3) - f(3))) + " off from the exact value")
````

## Exercise 6
A well known approximation, when $x$ is small, is $\cos(x)\simeq 1-x^2/2$. 
1. Bound the error of the approximation of $\sin(x)$ by $x$ for $-\pi/4\leq x \leq \pi/4$ using the plot. This means the maximum of $|\cos(x)-(1-x^2/2)|$ when $x$ is $-\pi/4\leq x \leq \pi/4$.
2. Bound the error using the formula.
3. Compare.

````python
pi = np.pi 
cos = np.cos
x = np.linspace(-pi / 4, pi / 4, 50)

def f(x):
  return cos(x)

def P(x):
  return 1 - x**2 / 2

plt.plot(x, f(x), label = "f(x)")
plt.plot(x, P(x), label = "P(x)")
plt.legend()
plt.show()

plt.plot(x, np.abs((f(x) - P(x))), label = "error: cos(x)  (1  (x^2)/2)")
plt.legend()
plt.show()

print("Max of approximate function g(x) : " + str(P(pi / 4)))
print("Max exact function f(x) : " + str(f(pi / 4)))
print("The error between the approximation an exact function is " + str(np.abs(P(pi / 4) - f(pi / 4))))
````
